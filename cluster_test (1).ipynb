```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('k-mean test.csv')
```

```python
data.head()
```

```python
happy = data['happyScore']
income = data['avg_income']
ineq = data['income_inequality']
gdp = data['GDP']
```

```python
plt.xlabel("income($)")
plt.ylabel("happiness")
plt.scatter(income,happy)

```

```python
data.sort_values('avg_income', inplace=True, ascending=False)
data
```

```python
richest = data[ data['avg_income'] > 15000]
richest
```

```python
richest.iloc[:-4,:-2]
```

```python
rich_mean = richest['avg_income'].mean()
all_mean = data['avg_income'].mean()

print(all_mean,  rich_mean)
```

```python

plt.xlabel('income')
plt.ylabel('happiness')

plt.scatter(richest['avg_income'], richest['happyScore'])
#labeling points{first} in the graph
plt.text(richest.iloc[0]['avg_income'],
         richest.iloc[0]['happyScore'],
         richest.iloc[0]['country'])
```

```python
plt.scatter(richest['avg_income'], richest['happyScore'])
#labeling points{last} in the graph
plt.text(richest.iloc[-1]['avg_income'],
         richest.iloc[-1]['happyScore'],
         richest.iloc[-1]['country'])
```

```python
#iterating over rows
for k,row in richest.iterrows():
    print(row['country'])
```

```python
#applying the iteration function to label every point in the graph
plt.scatter(richest['avg_income'], richest['happyScore'])
for k,row in richest.iterrows():
    plt.text(row['avg_income'],
             row['happyScore'],
             row['country'])
```

```python
#figuring out the use of intertuples(i.e for iterarting through index)
for row in richest.itertuples(index=True, name='Country'):
    print(row)
             
```

```python
#adding a third factor to properly understand the graph
plt.xlabel("income($)")
plt.ylabel("happiness")
plt.scatter(income,happy, s= ineq*8, alpha = 0.25)

```

```python
plt.xlabel("income($)")
plt.ylabel("happiness")
plt.scatter(income,happy, s= gdp*15, alpha=0.25)
```

```python
from sklearn.cluster import KMeans

happy
```

```python
#concatenate the income and happy values into a column
#using pandas func
income_happy = pd.concat([income, happy], axis = 1)
income_happy
```

```python
#concat using numpy function
import numpy as np
income_ha = np.column_stack((income, happy))
income_ha
```

```python
#cluster functions
km_res = KMeans(n_clusters=3).fit(income_happy)
km_res.cluster_centers_
```

```python
#cluster functions
km_ress = KMeans(n_clusters=3).fit(income_ha)
km_ress.cluster_centers_
#turns out that the functions from pandas and numpy give the same results
```

```python
#to plot the clusters on the graph
plt.scatter(income, happy)
plt.text(data.iloc[0]['avg_income'],
         data.iloc[0]['happyScore'],
         data.iloc[0]['country'])
plt.text(data.iloc[-1]['avg_income'],
         data.iloc[-1]['happyScore'],
         data.iloc[-1]['country'])
plt.text(data.iloc[-2]['avg_income'],
         data.iloc[-2]['happyScore'],
         data.iloc[-2]['country'])
plt.text(data.iloc[1]['avg_income'],
         data.iloc[1]['happyScore'],
         data.iloc[1]['country'])
clusters = km_res.cluster_centers_
plt.scatter(clusters[:,0], clusters[:,1], s= 1000)
```
